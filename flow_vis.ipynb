```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
feature_df = pd.read_csv('data/packet_rand_100%.csv')
plt.figure()
feature_df.hist(alpha=0.7,column='tot_pkt',bins=range(0,200,1))
plt.title('Ashkan - Packet random 100% sampling')
plt.show()
print('Min packet/flow:',feature_df['tot_pkt'].min())
print('Max packet/flow:',feature_df['tot_pkt'].max())
print('Stddev packet/flow:',feature_df['tot_pkt'].std())
print('Average packet/flow:',feature_df['tot_pkt'].mean())
print('Flow count:',feature_df['tot_pkt'].count())
```
