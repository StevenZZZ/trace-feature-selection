```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
LOCAL_PATH = 'data'
BENIGH_FLOW_CSV_FILE = 'packet_rand_{}%.csv'
DDOS_FLOW_CSV_FILE = 'ddos_pkt.csv'
```

```python
def show_packet_num_figure(filename,column_name,title,bin_range):
    feature_df = pd.read_csv(filename)
    plt.figure()
    feature_df.hist(alpha=0.7,column=column_name,bins=bin_range)
    plt.title(title)
    plt.show()
    print('Total packet number:', feature_df[column_name].sum())
    print('Min packet number of all flows:',feature_df[column_name].min())
    print('Max packet number of all flows',feature_df[column_name].max())
    print('Stddev packet number of all flows:',feature_df[column_name].std())
    print('Average packet number of all flows:',feature_df[column_name].mean())
    print('Flow count:',feature_df[column_name].count())
```

```python
# packet number statistics
for sampling_perc in [20,40,60,80,100]:
    show_packet_num_figure(LOCAL_PATH + '/' + BENIGH_FLOW_CSV_FILE.format(sampling_perc),'tot_pkt','Ashkan - Packet random {}% sampling - Packet Number'.format(sampling_perc),range(0,200,1))
```

```python
def show_packet_len_figure(filename,column_name,title,bin_range):
    feature_df = pd.read_csv(filename)
    plt.figure()
    feature_df.hist(alpha=0.7,column=column_name,bins=bin_range)
    plt.title(title)
    plt.show()
    print('Packet length(in bytes):', feature_df[column_name].sum())
    print('Min packet length(in bytes):',feature_df[column_name].min())
    print('Max packet length(in bytes):',feature_df[column_name].max())
    print('Stddev packet length(in bytes)',feature_df[column_name].std())
    print('Average packet length(in bytes):',feature_df[column_name].mean())
    print('Flow count:',feature_df[column_name].count())
```

```python
# packet number statistics
for sampling_perc in [20,40,60,80,100]:
    show_packet_len_figure(LOCAL_PATH + '/' + BENIGH_FLOW_CSV_FILE.format(sampling_perc),'tot_byte','Ashkan - Packet random {}% sampling - Packet Length'.format(sampling_perc),range(0,20000,100))
```
