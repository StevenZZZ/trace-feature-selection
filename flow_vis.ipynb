```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
LOCAL_PATH = 'data'
FLOW_CSV_FILE = 'packet_rand_{}%.csv'
```

```python
def show_packet_num_figure(perc,title):
    feature_df = pd.read_csv(LOCAL_PATH + '/' + FLOW_CSV_FILE.format(perc))
    plt.figure()
    feature_df.hist(alpha=0.7,column='tot_pkt',bins=range(0,200,1))
    plt.title('Ashkan - Packet random {} sampling - Packet Number'.format(title))
    plt.show()
    print('Total packet number:', feature_df['tot_pkt'].sum())
    print('Min packet number of all flows:',feature_df['tot_pkt'].min())
    print('Max packet number of all flows',feature_df['tot_pkt'].max())
    print('Stddev packet number of all flows:',feature_df['tot_pkt'].std())
    print('Average packet number of all flows:',feature_df['tot_pkt'].mean())
    print('Flow count:',feature_df['tot_pkt'].count())
```

```python
show_packet_num_figure(100,'no')
```

```python
show_packet_num_figure(80,'80%')
```

```python
show_packet_num_figure(60,'60%')
```

```python
show_packet_num_figure(40,'40%')
```

```python
show_packet_num_figure(20,'20%')
```

```python
def show_packet_len_figure(perc,title):
    feature_df = pd.read_csv(LOCAL_PATH + '/' + FLOW_CSV_FILE.format(perc))
    plt.figure()
    feature_df.hist(alpha=0.7,column='tot_byte',bins=range(0,20000,100))
    plt.title('Ashkan - Packet random {} sampling - Packet Length'.format(title))
    plt.show()
    print('Packet length(in bytes):', feature_df['tot_byte'].sum())
    print('Min packet length(in bytes):',feature_df['tot_byte'].min())
    print('Max packet length(in bytes):',feature_df['tot_byte'].max())
    print('Stddev packet length(in bytes)',feature_df['tot_byte'].std())
    print('Average packet length(in bytes):',feature_df['tot_byte'].mean())
    print('Flow count:',feature_df['tot_byte'].count())
```

```python
show_packet_len_figure(100,'no')
```

```python
show_packet_len_figure(80,'80%')
```

```python
show_packet_len_figure(60,'60%')
```

```python
show_packet_len_figure(40,'40%')
```

```python
show_packet_len_figure(20,'20%')
```
